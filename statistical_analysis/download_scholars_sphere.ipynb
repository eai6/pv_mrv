```python
import os
import requests
from bs4 import BeautifulSoup

# URL of the webpage containing the images
url = "https://scholarsphere.psu.edu/resources/08a985a4-d878-4fa9-b2f2-6060100580a9"

# Directory to save the downloaded images
output_directory = "downloaded_images"
os.makedirs(output_directory, exist_ok=True)

def download_images():
    try:
        # Send a GET request to the webpage
        response = requests.get(url)
        response.raise_for_status()

        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        

        # Find all image links
        links = soup.find_all('a', href=True)
        #image_links = [link['href'] for link in links if link['href'].endswith('.jpg')]
        image_links = []
        image_names = []
        for link in links:
            try:
                if link['class'][0] == 'matomo_download':
                    #print(link['href'])
                    image_links.append(link['href'])
                    image_names.append(link['title'])
            except:
                continue

        image_names = [name.split(" ")[1] for name in image_names]

        # Download each image
        for i in range(len(image_links)):
            image_link = image_links[i]
            image_name = image_names[i]
            print(image_link)
            image_url = image_link if image_link.startswith("http") else f"https://scholarsphere.psu.edu{image_link}"
            image_name = image_name #os.path.basename(image_url)
            image_path = os.path.join(output_directory, image_name)
            #image_path = image_path+'.jpg'

            # Download and save the image
            with requests.get(image_url, stream=True) as img_response:
                img_response.raise_for_status()
                with open(image_path, 'wb') as file:
                    for chunk in img_response.iter_content(chunk_size=8192):
                        file.write(chunk)

            print(f"Downloaded: {image_name}")

    except requests.exceptions.RequestException as e:
        print(f"Error during requests to {url}: {e}")

if __name__ == "__main__":
    download_images()
```
